```python
import librosa
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import dct
from sklearn.manifold import TSNE
import librosa.display

y, sr = librosa.load('dataset/test/test.wav', sr=None)  


# Set MFCC parameters
n_mfcc = 13
n_fft = 2048
hop_length = 512
win_length = 1024
n_mels = 40
fmin = 0
fmax = sr/2

# --- Manual MFCC Computation ---

pre_emphasis = 0.97
y_preem = np.append(y[0], y[1:] - pre_emphasis * y[:-1])

# Framing
num_frames = 1 + int((len(y_preem) - win_length)/hop_length)
frames = np.zeros((num_frames, win_length))
for i in range(num_frames):
    start = i * hop_length
    frames[i, :] = y_preem[start:start+win_length]

# Windowing
frames *= np.hamming(win_length)

# FFT -> Power Spectrum
mag_frames = np.absolute(np.fft.rfft(frames, n_fft))
pow_frames = (1.0 / n_fft) * (mag_frames ** 2)

# Mel filter banks
def hz_to_mel(hz):
    return 2595 * np.log10(1 + hz/700.)

def mel_to_hz(mel):
    return 700 * (10**(mel/2595.) - 1)

mel_points = np.linspace(hz_to_mel(fmin), hz_to_mel(fmax), n_mels + 2)
hz_points = mel_to_hz(mel_points)
bin_points = np.floor((n_fft+1)*hz_points/sr).astype(int)

fbank = np.zeros((n_mels, n_fft//2+1))
for m in range(1, n_mels+1):
    f_m_minus = bin_points[m-1]
    f_m = bin_points[m]
    f_m_plus = bin_points[m+1]

    for k in range(f_m_minus, f_m):
        fbank[m-1, k] = (k - f_m_minus) / (f_m - f_m_minus)
    for k in range(f_m, f_m_plus):
        fbank[m-1, k] = (f_m_plus - k) / (f_m_plus - f_m)

filter_banks = np.dot(pow_frames, fbank.T)
filter_banks = np.where(filter_banks == 0, np.finfo(float).eps, filter_banks)
log_mel_spect = np.log(filter_banks)
mfcc_manual = dct(log_mel_spect, type=2, axis=1, norm='ortho')[:, :n_mfcc]

# --- Librosa MFCC Computation ---
mfcc_librosa = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=n_mfcc,
                                    n_fft=n_fft, hop_length=hop_length,
                                    n_mels=n_mels, fmin=fmin, fmax=fmax,
                                    win_length=win_length, window='hamming')

mfcc_manual_t = mfcc_manual.T

# --- Apply t-SNE on MFCCs ---

# Let's use the manual MFCCs for demonstration.
tsne = TSNE(n_components=2, perplexity=30, random_state=42)
mfcc_2d = tsne.fit_transform(mfcc_manual)  # shape (num_frames, 2)

# --- Plot MFCCs and t-SNE ---
fig, axs = plt.subplots(3, 1, figsize=(10, 12))

# Plot MFCC - librosa
img0 = librosa.display.specshow(mfcc_librosa, x_axis='time', sr=sr, hop_length=hop_length, ax=axs[0])
axs[0].set_title('MFCC (librosa)')
axs[0].set_ylabel('MFCC Coefficients')
axs[0].set_xlabel('Time (frames)')
fig.colorbar(img0, ax=axs[0])

# Plot MFCC - Manual
img1 = librosa.display.specshow(mfcc_manual_t, x_axis='time', sr=sr, hop_length=hop_length, ax=axs[1])
axs[1].set_title('MFCC (Manual)')
axs[1].set_ylabel('MFCC Coefficients')
axs[1].set_xlabel('Time (frames)')
fig.colorbar(img1, ax=axs[1])


# Plot t-SNE of manual MFCC
axs[2].scatter(mfcc_2d[:,0], mfcc_2d[:,1], s=10, c='blue', alpha=0.7)
axs[2].set_title('t-SNE visualization of Manual MFCCs')
axs[2].set_xlabel('Dimension 1')
axs[2].set_ylabel('Dimension 2')
axs[2].grid(True)

plt.tight_layout()
plt.show()
```
